### I am modelling a delayed paired association task with RNNs. The network receives either of two sample inputs A or B and after a delay period either of two test inputs C or D. The networks needs to learn to give a positive readout when A is followed by C and B by D. This can be viewed in a simple toy model as a rotation in a 2D plane (A vs B axis and Response No response axis). In this model, test odors rotate the activity by plus or minus 45°. How can I simply explain that model?

A succinct way to picture the “two‐dimensional rotation” idea is this:

1.  First, imagine that the network’s activity in response to the sample
    (A vs. B) is represented somewhere along a horizontal axis in a 2D
    plane.
2.  The task then requires either a “respond” or “don’t respond”
    outcome, which can be placed along the vertical axis.
3.  When the test stimulus arrives (C or D), it shifts the internal
    state by a fixed rotation in that plane. In one case (say A followed
    by C), the rotation points the state into the “respond” quadrant; in
    the other case (A followed by D), the rotation points the state into
    the “no‐respond” quadrant.

In other words, each test stimulus nudges the internal representation
(which initially distinguishes A vs. B on one axis) by ±45 degrees so
that the new direction either lands in a region corresponding to
“positive” (respond) or “negative” (don’t respond). You can think of the
network learning to rotate its internal state in just the right way so
that the correct sample‐test pair combination ends up in the “respond”
region, while an incorrect sample‐test pair remains in the “no‐respond”
region.

### What I want to explain is that if you are on either side of the plane (right side encoding A and left side encoding B) then the only way to perform the task (go up for AC and BD and go down for AD and BC) is with a rotation

One way to see why you “must” use a rotation in the 2D plane is simply
to check that A and B start on opposite sides (e.g., right vs. left),
and the task demands that each of those two starting points must be sent
“up” under one test cue (C or D) and “down” under the other.
Geometrically, a uniform shift (translation) would push both A and B in
the same direction and could never separate them into “up” vs. “down.”
Similarly, a reflection or a simple scaling does not solve the problem
for both A and B at once. By contrast, a ±45° rotation centered at the
origin naturally takes “A on the right” to either “up” or “down” and
likewise takes “B on the left” to the opposite choice. Indeed, if you
place A at (1, 0) and B at (−1, 0), then rotating by +45° sends A up but
sends B down; rotating by −45° does the reverse. Thus, a rotation is the
simplest linear transformation that maps (A→up, B→down) or (A→down,
B→up) depending on the test cue.

Because pure translations would shift both A and B equally (never
separating them into up/down) and scalings or reflections also fail to

and B → up under the other test odor.

That’s the essence of why “rotation” in a 2D plane is the simplest
schematic explanation for this paired association task.

### What counter arguments could you find against rotations in the activity plane?

Below are some counterarguments or caveats to the idea of a strict
rotation in a 2D activity plane:

1.  The network might use more complex transformations. • Real recurrent
    networks often work in higher‐dimensional state spaces and can
    employ nonlinear dynamics. What looks like a neat geometric rotation
    in 2D could actually be just a 2D projection of more complicated
    activity patterns.

2.  A pure rotation is not the only linear mapping. • Other linear
    transformations (e.g., shear transformations, reflections, or
    combinations thereof) could accomplish the “map A to ‘up’, B to
    ‘down’” scheme if you allow for a suitable readout boundary.
    Rotation is simply one particularly intuitive choice.

3.  Translations or shifts could matter in higher‐dimensional state
    spaces. • Even if a purely 2D toy model suggests rotation is the
    simplest visualization, in a high‐dimensional network it could be a
    combined shift + rotation + nonlinearity.

4.  The network readout need not be purely “up vs. down.” • In practice,
    the network might use a more general decision boundary. A 2D
    rotation is a nice schematic but not necessarily the literal way
    real neurons or artificial RNNs implement the mapping.

5.  Noise, variability, and heterogeneity in real networks. • Actual
    neural or large RNN activity can be messy—small deviations from a
    perfect rotation might still solve the task. The clean rotation
    story is an idealized explanation rather than a strict requirement.
